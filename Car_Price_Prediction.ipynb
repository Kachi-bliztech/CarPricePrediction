Importing the Dependencies
***Note: Data needs further processing to get better result********

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

Data Collection and Processing

```python
# loading the data from csv file to pandas dataframe
car_dataset = pd.read_csv('CAR DETAILS FROM CAR DEKHO.csv')
```

```python
car_dataset.head()
```

```python
# checking the number of rows and columns
car_dataset.shape
```

```python
# getting some information about the dataset
car_dataset.info()
```

```python
# checking the number of missing values
car_dataset.isnull().sum()
```

```python
# checking the distribution of categorical data
print(car_dataset.fuel.value_counts())
print(car_dataset.seller_type.value_counts())
print(car_dataset.transmission.value_counts())
```

Encoding the Categorical Data

```python
# encoding "Fuel_Type" Column
car_dataset.replace({'fuel':{'Petrol':0,'Diesel':1,'CNG':2, 'LPG':3, 'Electric':4 }},inplace=True)

# encoding "Seller_Type" Column
car_dataset.replace({'seller_type':{'Dealer':0,'Individual':1, 'Trustmark Dealer':2}},inplace=True)

# encoding "Transmission" Column
car_dataset.replace({'transmission':{'Manual':0,'Automatic':1}},inplace=True)
```

```python
print(car_dataset.owner.value_counts())
```

```python
car_dataset.replace({'owner':{'First Owner':0,'Second Owner':1,'Third Owner':2,'Fourth & Above Owner':3,'Test Drive Car':1  }},inplace=True)
```

```python
print(car_dataset.owner.value_counts())
```

```python
car_dataset.head()
```

```python
car_dataset = car_dataset['selling_price'].round(2)
```

```python
car_dataset.head()
```

Splitting the data and Target

```python
X = car_dataset.drop(['name','selling_price'],axis=1)
Y = car_dataset['selling_price']
```

```python
X
```

```python
Y
```

Splitting Training and Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, random_state=2)
```

Model Training

1. Linear Regression

```python
# loading the linear regression model
lin_reg_model = LinearRegression()
```

```python
lin_reg_model.fit(X_train,Y_train)
```

Model Evaluation

```python
# prediction on Training data
training_data_prediction = lin_reg_model.predict(X_train)
```

```python
# R squared Error
error_score = metrics.r2_score(Y_train, training_data_prediction)
print("R squared Error : ", error_score)
```

Visualize the actual prices and Predicted prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```

```python
# prediction on Training data
test_data_prediction = lin_reg_model.predict(X_test)
```

```python
# R squared Error
error_score = metrics.r2_score(Y_test, test_data_prediction)
print("R squared Error : ", error_score)
```

```python
plt.scatter(Y_test, test_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```
